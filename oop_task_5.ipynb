1. Write a Python programme to create the parent class shape, derive the subclasses circle,
square, and rectangle from the class shape, and demonstrate polymorphism in inheritance
using a member function area ().

```python
class Shape:
  def area(self):
    pass


#Square
class Square(Shape):
  def __init__ (self, side):
    super().__init__()
    self.side = side
  
  def area(self) :
    return self.side * self.side
    pass     

#Circle
class Circle(Shape):
  pi = 3.14
  def __init__ (self, radius):
    super().__init__()
    self.radius = radius

  def area(self) :
    return round(Circle.pi * (self.radius ** 2), 2)
    pass  

#Rectangle
class Rectangle(Shape):
  def __init__ (self, length, breadth):
    super().__init__()
    self.length = length
    self.breadth = breadth
 
  def area(self) :
    return self.length * self.breadth
    pass      


r1 = Rectangle (50, 20)
print("Area of rectangle :", r1.area())

c1 = Circle(12)
print("Area of circle :", format(c1.area(), "0.2f"))

s1 = Square (15) 
print("Area of square :", s1.area())
```

2. The family has decided to travel by road from Dharwad to Belagavi, and they have
chosen to estimate the cost before starting on this journey. Write python code to create
the parent class Travel with member functions include the number of passengers, the
distance travelled, and the mode of transportation. Derive two subclasses, train and bus,
from the parent class, each with a member function that calculates the cost of trip. Also,
keep the numbers of passengers as a private member of the parent class and try to access
it through the child class to estimate costs.
[Note: Assume cost of travelling/person through train = 60 rs, cost of travelling /person
through bus= 100 rs].

```python

class Travel:

  def __init__(self, passengers):
    self.__passengers=passengers

  @property       #decorator
  def passengers(self):
    return self.__passengers

  def distance(self):
    pass

  def mode(self):
    pass

class train(Travel):
  def cost_of_transport(self):
    print(self.passengers*60)

class bus(Travel):
  def cost_of_transport(self):
    print(self.passengers*100)

mode1 = train(10)
print("The cost of trip by train =")
mode1.cost_of_transport()

mode2 = bus(10)
print("The cost of trip by bus =")
mode2.cost_of_transport()


seconds_in_a_day = 24 * 60 * 60
seconds_in_a_day

seconds_in_a_week = 7 * seconds_in_a_day
seconds_in_a_week

class Travel:

  def __init__(self, passengers):
    self.__passengers=passengers

  @property       #decorator
  def passengers(self):
    return self.__passengers

  def distance(self):
    pass

  def mode(self):
    pass

class train(Travel):
  def cost_of_transport(self):
    print(self.passengers*60)

class bus(Travel):
  def cost_of_transport(self):
    print(self.passengers*100)

mode1 = train(10)
print("The cost of trip by train =")
mode1.cost_of_transport()

mode2 = bus(10)
print("The cost of trip by bus =")
mode2.cost_of_transport()
```
